```python
from google.colab import userdata
```

```python
from huggingface_hub import InferenceClient


client = InferenceClient(model="gpt2", token=userdata.get('commonwrite'))

prompt = "what does tesla do?"
response = client.text_generation(prompt, max_new_tokens=200)
print(response)
```

```python
# question answering

qaclient = InferenceClient(model="deepset/bert-large-uncased-whole-word-masking-squad2", token=userdata.get('commonwrite'))

experiment = qaclient.question_answering(question="who is virat kohli?",context="On 18 December 2006, Kohli experienced the loss of his father due to a cerebral attack.[13][16] During his childhood, his father played a crucial role in supporting his cricket training. Kohli has credited his father as the one who drove him to practice every day. He has expressed his feelings of missing his father's presence at times.[10][17] According to his mother, Kohli's demeanour shifted noticeably after his father's death. He took cricket with newfound seriousness, prioritizing playing time and dedicating himself fully to the sport.[13] Kohli's family resided in Meera Bagh, Paschim Vihar until the year 2015, after which they relocated to Gurgaon.[18]")

print(experiment)
```

```python
# text to image

imgclient = InferenceClient(model="black-forest-labs/FLUX.1-dev", token=userdata.get('commonwrite'))

prompt = "Virat kohli playing football"
result = imgclient.text_to_image(prompt)

result
```
